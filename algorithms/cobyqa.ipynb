```python
import numpy as np
from scipy.optimize import minimize

def ackley(x):
    x = np.asarray(x)
    d = len(x)
    
    term1 = -20 * np.exp(-0.2 * np.sqrt(np.sum(x**2) / d))
    term2 = -np.exp(np.sum(np.cos(2 * np.pi * x)) / d)
    
    return term1 + term2 + 20 + np.e

dim = 20
bounds = [(-32.768, 32.768)] * dim

np.random.seed(27)
x0 = np.random.uniform(-5, 5, size=dim)


constraints = []

for i in range(dim):
    lb, ub = bounds[i]
    
    constraints.append({
        'type': 'ineq',
        'fun': lambda x, i=i, lb=lb: x[i] - lb
    })
    
    constraints.append({
        'type': 'ineq',
        'fun': lambda x, i=i, ub=ub: ub - x[i]
    })


res = minimize(
    ackley,
    x0,
    method='COBYQA',
    bounds=bounds,
    options={
        'maxiter': 5000,
        'rhobeg': 1.0,
        'tol': 1e-6,
        'disp': True
    }
)

print("Best x:", res.x)
print("Best f(x):", res.fun)
print("Iterations:", res.nfev)
print("Success:", res.success)
```
