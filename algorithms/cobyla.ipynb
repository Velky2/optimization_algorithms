### COBYLA

### 1. Visão geral

O COBYLA, proposto em 1994 por Powell, se propõe a resolver problemas do tipo:
$$
min_{x \in \mathbb{R}} f(x); \space c_i(x) \geq 0; \space i=1,2,...,m
$$
Com f e c funções blackbox.

### 2. Hiperplano

O objetivo do COBYLA é aproximar um hiperplano que melhor aproxime os n pontos disponíveis.

Para isso, o COBYLA objetiva encontrar $f'(x^{(k)}) = f(x^{(k)})$ para todo k, e $f'(x) = a_0 + a^Tx$

Em notação matricial, o algoritmo estará resolvendo o problema:
$$
A =
\begin{bmatrix}
1 & (x^{(0)}_1) & (x^{(0)}_2) & \cdots & (x^{(0)}_n) \\
1 & (x^{(1)}_1) & (x^{(1)}_2) & \cdots & (x^{(1)}_n) \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & (x^{(n)}_1) & (x^{(n)}_2) & \cdots & (x^{(n)}_n)
\end{bmatrix}
$$
ou
$$
A =
\begin{bmatrix}
1 & (x^{(0)})^T \\
1 & (x^{(1)})^T \\
\vdots & \vdots \\
1 & (x^{(n)})^T
\end{bmatrix}
$$
$$
\theta =
\begin{bmatrix}
a_0 \\
a_1 \\
a_2 \\
\vdots \\
a_n
\end{bmatrix}
$$
$$
f =
\begin{bmatrix}
f(x^{(0)}) \\
f(x^{(1)}) \\
\vdots \\
f(x^{(n)})
\end{bmatrix}
$$

O objetivo é resolver $A\theta = f$ ou, $\theta = A^{-1}f$

### 3. Trust Region Linear

Tendo-se o hiperplano, 

Primeiramente, façamos uma otimização usando o COBYLA do scipy

```python
import numpy as np
from scipy.optimize import minimize

def ackley(x):
    x = np.asarray(x)
    d = len(x)
    
    term1 = -20 * np.exp(-0.2 * np.sqrt(np.sum(x**2) / d))
    term2 = -np.exp(np.sum(np.cos(2 * np.pi * x)) / d)
    
    return term1 + term2 + 20 + np.e

dim = 20
bounds = [(-32.768, 32.768)] * dim

np.random.seed(27)
x0 = np.random.uniform(-5, 5, size=dim)


constraints = []

for i in range(dim):
    lb, ub = bounds[i]
    
    constraints.append({
        'type': 'ineq',
        'fun': lambda x, i=i, lb=lb: x[i] - lb
    })
    
    constraints.append({
        'type': 'ineq',
        'fun': lambda x, i=i, ub=ub: ub - x[i]
    })


res = minimize(
    ackley,
    x0,
    method='COBYLA',
    constraints=constraints,
    options={
        'maxiter': 5000,
        'rhobeg': 1.0,
        'tol': 1e-6,
        'disp': True
    }
)

print("Best x:", res.x)
print("Best f(x):", res.fun)
print("Iterations:", res.nfev)
print("Success:", res.success)
```

```python
import numpy as np
from scipy.optimize import minimize

def ackley(x):
    x = np.asarray(x)
    d = len(x)
    
    term1 = -20 * np.exp(-0.2 * np.sqrt(np.sum(x**2) / d))
    term2 = -np.exp(np.sum(np.cos(2 * np.pi * x)) / d)
    
    return term1 + term2 + 20 + np.e

dim = 20
bounds = [(-32.768, 32.768)] * dim
results = []

for i in range(50):
    np.random.seed(i)
    x0 = np.random.uniform(-5, 5, size=dim)


    constraints = []

    for i in range(dim):
        lb, ub = bounds[i]
        
        constraints.append({
            'type': 'ineq',
            'fun': lambda x, i=i, lb=lb: x[i] - lb
        })
        
        constraints.append({
            'type': 'ineq',
            'fun': lambda x, i=i, ub=ub: ub - x[i]
        })


    res = minimize(
        ackley,
        x0,
        method='COBYLA',
        constraints=constraints,
        options={
            'maxiter': 5000,
            'rhobeg': 1.0,
            'tol': 1e-6,
            'disp': True
        }
    )
    print(res.fun)
    results.append(res.fun)

print("Média:")
print(sum(results)/50)
```
